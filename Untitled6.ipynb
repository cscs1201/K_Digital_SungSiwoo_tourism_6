```python
def Pandas_NumPy_Transformation(data): # 파이썬 배열을 집어 넣는다.
    PNT_Numpy_Array=([])
    for Inside_pandas in data:
        PNT_Numpy_Array = np.append(PNT_Numpy_Array,np.array(Inside_pandas))
        
    return PNT_Numpy_Array
```
